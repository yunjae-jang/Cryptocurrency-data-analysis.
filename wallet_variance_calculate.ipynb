```python
import matplotlib.pyplot as plt
import numpy as np

# 1. 기본 스타일 설정
plt.style.use('default')
plt.rcParams['figure.figsize'] = (16, 7)
plt.rcParams['font.size'] = 12

# 3. 그래프 그리기
fig, ax1 = plt.subplots()
wallet.plot(x='coin', y='health_score' ,kind='bar', ax=ax1,label='_nolegend_',color="purple")
ax1.tick_params(axis='both', direction='in')
plt.xticks(rotation=45)

ax2 = ax1.twinx()
wallet.plot('coin', 'ratio', marker='o', color='orange', markersize=5, linewidth=2, alpha=0.9, label='wallet_var', ax=ax2)
ax2.set_ylabel('Health_score')
ax2.tick_params(axis='y', direction='in')
```
