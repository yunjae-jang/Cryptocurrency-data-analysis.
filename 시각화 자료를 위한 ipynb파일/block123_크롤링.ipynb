```python
driver = webdriver.Chrome('D:/temp/chromedriver.exe') 
url = 'https://www.block123.com/account/login/'
driver.get(url)
driver.maximize_window() #웹을 가장 크게 여는 함수다. 
action = ActionChains(driver)

# Block123 로그인 
driver.find_element_by_css_selector('#id_login').send_keys('psps4755@naver.com')
driver.find_element_by_css_selector('#id_password').send_keys('36330903ps@')
driver.find_element_by_css_selector('#id_password').send_keys(Keys.ENTER)

good_dict = {}
# Block123 검색하기 
for coin in coinlist : 
    driver.find_element_by_css_selector('.search-text-input').send_keys(coin)
    driver.find_element_by_css_selector('.search-text-input').send_keys(Keys.ENTER)
    time.sleep(1)

    # 검색어 첫번째 클릭하기 
    driver.find_element_by_css_selector('.name-wrapper').click()
    time.sleep(1)

    # 투자기업 이름 따오기
    invest = []
    for name in driver.find_elements_by_css_selector('.name'):
        invest.append(name.text)
        good_dict[coin] = invest[:]      
```
