```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = df.drop(columns=['Unnamed: 0', 'Price Change', '24h Low', '24h High', 'Fully Diluted Market Cap',
       'Yesterdays Low', 'Yesterdays High', 'Yesterdays Open',
       'Yesterdays Close', 'Yesterdays Change', 'Yesterdays Volume', '7d Low',
       '7d High', '30d Low', '30d High', '90d Low', '90d High', '52 Week Low',
       '52 Week High', 'All Time High', 'All Time Low', 'ROI', 'Crirulating Supply', 'Total Supply',
       'Max Supply', 'Amount'])
```

```python
df = df.iloc[[0,1,2,3,5,6,11,12,14,16,20]]
```

```python
df=df.reset_index()
df=df.drop(columns='index')
```

```python
df
```

```python
df.columns
```

```python
df.columns=['코인명', '가격(KRW)', '거래량(KRW)', '시총 대비 1일 거래량',
       '시장 도미넌스(장악력)(%)', '시장 순위', '시총', '출시 시점대비 투자 수익(배)',
       '총 공급량 대비 순환 공급량', 'BEST 20의 코인 보유율(%)',
       '펀더멘털 점수', '투자한 회사', '투자한 회사 수', '플랫폼',
       '카테고리', '이자율(%)']

```

```python
df.head()
```

```python
df['시장 도미넌스(장악력)(%)'] = df['시장 도미넌스(장악력)(%)'].str.replace('%', '')
df['BEST 20의 코인 보유율(%)'] = df['BEST 20의 코인 보유율(%)'].str.replace('%', '')
df['시장 도미넌스(장악력)(%)'] = df['시장 도미넌스(장악력)(%)'].astype(float)
df.head()
```

```python
# 코인별 시장 도미넌스(장악력)(%)

plt.figure(figsize=(15,3))
sns.barplot(data=df, x='코인명', y='시장 도미넌스(장악력)(%)', ci=None,color="#692498")
```

```python
df.info()
```

```python
# 코인별 거래량(KRW)

pd.options.display.float_format = '{:.14f}'.format
plt.figure(figsize=(15,3))
sns.barplot(data=df, x='코인명', y='거래량(KRW)', ci=None)
```

```python
# 코인별 시총

pd.options.display.float_format = '{:.14f}'.format
plt.figure(figsize=(15,3))
sns.barplot(data=df, x='코인명', y='시총', ci=None)
```

```python
# 코인별 출시 시점대비 투자 수익(배)

pd.options.display.float_format = '{:.14f}'.format
plt.figure(figsize=(15,3))
sns.barplot(data=df, x='코인명', y='출시 시점대비 투자 수익(배)', ci=None,color="#692498")
```

```python
# 코인별 시총 대비 1일 거래량

pd.options.display.float_format = '{:.14f}'.format
plt.figure(figsize=(15,3))
sns.barplot(data=df, x='코인명', y='시총 대비 1일 거래량', ci=None)
```
